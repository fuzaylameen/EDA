```python
import pandas as pd
import numpy as np

# -----------------------------
# 1. Load Datasets
# -----------------------------
customer_df = pd.read_csv("Customer_Info.csv")
transaction_df = pd.read_csv("Transactions.csv")

print("Original Customer Data:")
print(customer_df)

print("\nOriginal Transaction Data:")
print(transaction_df)
```

```python
# -----------------------------
# 2. Remove Duplicate Rows
# -----------------------------
customer_df = customer_df.drop_duplicates()
transaction_df = transaction_df.drop_duplicates()

print("\nAfter Removing Duplicates:")
print(customer_df)
```

```python
# -----------------------------
# 3. Replace Customer_ID with Generic Numeric IDs
# -----------------------------
# Create mapping
customer_id_mapping = {old_id: new_id for new_id, old_id in enumerate(customer_df['Customer_ID'].unique(), start=1)}

# Apply mapping
customer_df['Customer_ID'] = customer_df['Customer_ID'].map(customer_id_mapping)
transaction_df['Customer_ID'] = transaction_df['Customer_ID'].map(customer_id_mapping)

print("\nAfter Replacing Customer_ID with Numeric IDs:")
print(customer_df)
```

```python
# -----------------------------
# 4. Format Phone Numbers Consistently
# -----------------------------
# Convert phone numbers to string and ensure 10 digits
customer_df['Phone_Number'] = customer_df['Phone_Number'].astype(str).str.replace(r'\D', '', regex=True)
customer_df['Phone_Number'] = customer_df['Phone_Number'].str.zfill(10)
```

```python
# -----------------------------
# 5. Fill Missing Loyalty Points with Average
# -----------------------------
avg_loyalty = customer_df['Loyalty_Points'].mean()
customer_df['Loyalty_Points'] = customer_df['Loyalty_Points'].fillna(avg_loyalty)

print("\nAfter Formatting Phone & Filling Loyalty:")
print(customer_df)
```

```python
# -----------------------------
# 6. Check for Negative Quantities
# -----------------------------
invalid_quantity = transaction_df[transaction_df['Quantity'] < 0]

if not invalid_quantity.empty:
    print("\nInvalid Negative Quantities Found:")
    print(invalid_quantity)
else:
    print("\nNo Negative Quantities Found")
```

```python
# -----------------------------
# 7. Check for Invalid Payment Methods
# -----------------------------
valid_payment_methods = ["Credit Card", "Cash", "UPI", "Debit Card"]

invalid_payments = transaction_df[~transaction_df['Payment_Method'].isin(valid_payment_methods)]

if not invalid_payments.empty:
    print("\nInvalid Payment Methods Found:")
    print(invalid_payments)
else:
    print("\nNo Invalid Payment Methods Found")
```

```python
# -----------------------------
# 8. Derive Total_Amount Column
# -----------------------------
transaction_df['Total_Amount'] = transaction_df['Quantity'] * transaction_df['Price_Per_Unit']

print("\nTransaction Data with Total_Amount:")
print(transaction_df)
```

```python
# -----------------------------
# 9. Aggregate Total Purchase Amount by Customer_ID
# -----------------------------
total_purchase = transaction_df.groupby('Customer_ID')['Total_Amount'].sum().reset_index()
total_purchase.rename(columns={'Total_Amount': 'Total_Purchase_Amount'}, inplace=True)

print("\nTotal Purchase Amount per Customer:")
print(total_purchase)
```
