```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# -----------------------------
# Create Dataset
# -----------------------------
data = {
    "Order_ID": [1001,1002,1003,1004,1005,1006,1007,1008,1009,1010,1011,1012,1013,1014,1015],
    "Product": ["Laptop","Mobile","Laptop","Tablet","Mobile","Laptop","Tablet","Mobile","Laptop","Tablet","Mobile","Laptop","Mobile","Tablet","Laptop"],
    "Region": ["North","South","East","West","North","South","East","West","North","South","East","West","North","South","East"],
    "Sales_Amount": [85000,42000,90000,30000,45000,78000,35000,41000,92000,33000,46000,86000,40000,32000,87000],
    "Units_Sold": [5,10,6,8,12,4,10,11,7,9,13,5,9,10,6],
    "Month": ["January","January","February","February","March","March","April","April","May","May","June","June","July","July","August"],
    "Customer_Rating": [4.2,3.8,4.5,3.6,4.0,4.4,3.9,4.1,4.7,3.7,4.3,4.6,4.0,3.8,4.4]
}

df = pd.DataFrame(data)

# -----------------------------
# 1. Average Sales Amount by Region (Bar Plot)
# -----------------------------
avg_sales = df.groupby("Region")["Sales_Amount"].mean()

plt.figure()
avg_sales.plot(kind="bar")
plt.title("Average Sales Amount by Region")
plt.xlabel("Region")
plt.ylabel("Average Sales Amount")
plt.show()

# -----------------------------
# 2. Monthly Sales Trend (Line Plot)
# -----------------------------
monthly_sales = df.groupby("Month")["Sales_Amount"].sum()
monthly_sales = monthly_sales.reindex(
    ["January","February","March","April","May","June","July","August"]
)

plt.figure()
plt.plot(monthly_sales.index, monthly_sales.values, marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales Amount")
plt.xticks(rotation=45)
plt.show()

# -----------------------------
# 3. Relationship between Units Sold and Sales Amount (Scatter Plot)
# -----------------------------
plt.figure()
plt.scatter(df["Units_Sold"], df["Sales_Amount"])
plt.title("Units Sold vs Sales Amount")
plt.xlabel("Units Sold")
plt.ylabel("Sales Amount")
plt.show()

# -----------------------------
# 4. Distribution of Customer Rating across Products (Box Plot)
# -----------------------------
plt.figure()
sns.boxplot(x="Product", y="Customer_Rating", data=df)
plt.title("Customer Rating Distribution by Product")
plt.show()

# -----------------------------
# 5. Correlation Heatmap
# -----------------------------
corr = df[["Sales_Amount","Units_Sold","Customer_Rating"]].corr()

plt.figure()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation between Sales Amount, Units Sold, and Customer Rating")
plt.show()
```
