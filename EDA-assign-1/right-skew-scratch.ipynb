```python
# Dataset
data = [10, 12, 13, 14, 15, 18, 40]

n = len(data)

# 1. Mean
mean = sum(data) / n

# 2. Median
sorted_data = sorted(data)
if n % 2 == 0:
    median = (sorted_data[n//2 - 1] + sorted_data[n//2]) / 2
else:
    median = sorted_data[n//2]

# 3. Mode
# Count frequency of each element
freq = {}
for num in data:
    freq[num] = freq.get(num, 0) + 1

max_freq = max(freq.values())
modes = [k for k, v in freq.items() if v == max_freq]

# If all values are unique, no mode
if max_freq == 1:
    mode = None
else:
    mode = modes

# 4. Variance (sample variance)
squared_diff = [(x - mean) ** 2 for x in data]
variance = sum(squared_diff) / (n - 1)

# 5. Standard Deviation
std_dev = variance ** 0.5

# 6. Skewness (sample skewness)
skew_sum = sum(((x - mean) / std_dev) ** 3 for x in data)
skewness = (n / ((n - 1) * (n - 2))) * skew_sum

# Print results
print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
print("Variance:", variance)
print("Standard Deviation:", std_dev)
print("Skewness:", skewness)
```
