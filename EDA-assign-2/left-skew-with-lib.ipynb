```python
import numpy as np
import matplotlib.pyplot as plt

# -------------------- LEFT-SKEWED DATASET --------------------
data = np.array([30, 60, 70, 75, 80, 85, 90])

# -------------------- KARL PEARSON 2nd COEFFICIENT --------------------
def karl_pearson_skewness(x):
    mean = np.mean(x)
    median = np.median(x)
    std = np.std(x)
    return 3 * (mean - median) / std

# -------------------- ORIGINAL SKEWNESS --------------------
print("Original Dataset Skewness:")
original_skew = karl_pearson_skewness(data)
print("Original:", original_skew)

# -------------------- TRANSFORMATIONS --------------------
transformations = {
    "Original": data,
    "Min-Max": (data - np.min(data)) / (np.max(data) - np.min(data)),
    "Z-Score": (data - np.mean(data)) / np.std(data),
    "Robust": (data - np.median(data)) / (np.percentile(data, 75) - np.percentile(data, 25)),
    "Log": np.log(data),
    "Sqrt": np.sqrt(data)
}

# -------------------- COMPUTE SKEWNESS --------------------
skewness_values = {}

print("\nKarl Pearson’s 2nd Coefficient of Skewness:")
for name, values in transformations.items():
    skew = karl_pearson_skewness(values)
    skewness_values[name] = skew
    print(name, ":", skew)

# -------------------- LEAST SKEWNESS --------------------
least_skew = min(skewness_values, key=lambda x: abs(skewness_values[x]))
print("\nTransformation with least skewness:", least_skew)

# -------------------- BAR GRAPH --------------------
plt.figure()
plt.bar(skewness_values.keys(), skewness_values.values())
plt.ylabel("Skewness")
plt.title("Karl Pearson Skewness for Left-Skewed Data")
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```
